1. Obtén la sucursal que más vendió en la segunda quincena de Enero.

```python

import pandas as pd
import sqlite3 as sql 
from pym import tools as t

# 1. DISTINCT: se le aplica a una columna
# a. Valores unicos de esa columna
# b. Eliminar duplicados

# Selecciona TODAS las columnas de la tabla ventas
query = """
SELECT 
   Sucursal, 
   CAST(SUM(Total) AS INTEGER) AS Total_venta
FROM 
    ventas
WHERE 
---- Between lo utlilizamos para seleccionar valores de la columna fecha
--- que se encuentran entre "2025-01-16" y "2025-01-31"
    Fecha BETWEEN "2025-01-16" AND "2025-01-31"
GROUP BY 
    Sucursal
ORDER BY 
    Total_venta DESC
LIMIT 1
"""
t.consultar_bbdd(query)
```


2. ¿Que producto se vendió menos en la última semana de Enero? ¿En qué sucursal dicho producto
vendió menos?

```python
import pandas as pd
import sqlite3 as sql 
from pym import tools as t


# Selecciona la columna producto y suma todas las cantidades para después renombrarlo como Cantidad_total_producto
query = """
SELECT 
   Producto, 
   SUM(Cantidad) AS Cantidad_total_producto
FROM 
    ventas
WHERE 
---- Between lo utlilizamos para seleccionar valores de la columna fecha
--- que se encuentran entre "2025-01-25" y "2025-01-31"
    Fecha BETWEEN "2025-01-25" AND "2025-01-31"
GROUP BY 
    Producto
ORDER BY 
    Cantidad_total_producto ASC
    
LIMIT 1
"""
t.consultar_bbdd(query)
```

3. ¿Cuál fue el ticket promedio de ventas el 6 de enero en Polanco?

```python
import pandas as pd
import sqlite3 as sql 
from pym import tools as t


# Selecciona la columna total y saca promedio y lo renombramos como ticket promedio
# Esto se selecciona de la tabla ventas y se hace la seleccion solo de la sucursas polanco y  cuando la fecha
# es "2025-01-06" 
query = """
SELECT 
   AVG(Total) AS Ticket_promedio
FROM 
    ventas
WHERE 
    Fecha= "2025-01-06" 
    AND Sucursal="Polanco"

"""
t.consultar_bbdd(query)
```
